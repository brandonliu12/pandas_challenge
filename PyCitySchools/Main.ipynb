```python
# Import Dependencies
import pandas as pd
```

```python
csv_path = "../Resources/schools_complete.csv"
csv_path2 = "../Resources/students_complete.csv"
school_df = pd.read_csv(csv_path)
student_df = pd.read_csv(csv_path2)
student_df
school_df
schools = school_df["school_name"].nunique()
students = student_df["Student ID"].nunique()
```

```python
#District Summary
total_schools = len(school_df['school_name'].unique())
total_students = student_df['Student ID'].count()
total_budget = school_df['budget'].sum()
avg_math_score = student_df['math_score'].mean()
avg_reading_score = student_df['reading_score'].mean()
passing_math_percent = (student_df[student_df['math_score'] >= 70]['Student ID'].count() / total_students) * 100
passing_reading_percent = (student_df[student_df['reading_score'] >= 70]['Student ID'].count() / total_students) * 100
passing_overall_percent = (student_df[(student_df['math_score'] >= 70) & (student_df['reading_score'] >= 70)]['Student ID'].count() / total_students) * 100
results_dict = {
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [avg_math_score],
    'Average Reading Score': [avg_reading_score],
    '% Passing Math': [passing_math_percent],
    '% Passing Reading': [passing_reading_percent],
    '% Overall Passing': [passing_overall_percent]
}
district_results_df = pd.DataFrame(data=results_dict)
print(district_results_df)
```

```python
#school summary
merged_df = pd.merge(school_df,student_df,on=["school_name"],how="inner")
merged_df["passing_math"] = merged_df['math_score'] >= 70
merged_df["passing_reading"] = merged_df['reading_score'] >= 70
merged_df["passing_overall"] = (merged_df['math_score'] >= 70) & (student_df['reading_score'] >= 70)
grouped_df = merged_df.groupby('school_name')

# Calculate the desired metrics for each school
total_students = grouped_df['Student ID'].count()
total_budget = grouped_df['budget'].mean()
avg_math_score = grouped_df['math_score'].mean()
avg_reading_score = grouped_df['reading_score'].mean()
passing_math_percent = (merged_df[merged_df["passing_math"] == True].groupby('school_name')['Student ID'].count() / total_students)*100
passing_reading_percent = (merged_df[merged_df['passing_reading'] == True].groupby('school_name')['Student ID'].count() / total_students)*100
passing_overall_percent = (merged_df[merged_df['passing_overall'] == True].groupby('school_name')['Student ID'].count() / total_students)*100
type = grouped_df['type'].unique()

# Combine the metrics into a single DataFrame
results_dict = {
    'type': type,
    'Total Students': total_students,
    'Total Budget': total_budget,
    'Average Math Score': avg_math_score,
    'Average Reading Score': avg_reading_score,
    '% Passing Math': passing_math_percent,
    '% Passing Reading': passing_reading_percent,
    '% Overall Passing': passing_overall_percent
    
}
results_df = pd.DataFrame(data=results_dict)

# Print the results for each school
print(results_df)
```

```python
highest_passing_df = results_df.sort_values("% Overall Passing", ascending=False)
highest_passing_df.head(5)
```

```python
#Average Scores
grouped_df = merged_df.groupby(['school_name', 'grade'])
avg_math_score = grouped_df['math_score'].mean()
avg_reading_score = grouped_df['reading_score'].mean()
avg_math_score
avg_reading_score
```

```python
#Scores by spending

# create bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
results_df["average_budget"] = total_budget / total_students

# categorize spending based on the bins
results_df["spending_ranges (Per Student)"] = pd.cut(results_df["average_budget"], spending_bins, labels=labels)

# calculate mean scores per spending range
spending_math_scores = results_df.groupby(["spending_ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = results_df.groupby(["spending_ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = results_df.groupby(["spending_ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = results_df.groupby(["spending_ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = results_df.groupby(["spending_ranges (Per Student)"]).mean()["% Overall Passing"]

# create DataFrame with the above metrics
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# display the table
display(spending_summary)
```

```python
#Scores by school size

# create bins and labels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# categorize school size based on the bins
results_df["School Size"] = pd.cut(results_df["Total Students"], size_bins, labels=labels)

# calculate mean scores per school size
size_math_scores = results_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = results_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = results_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = results_df.groupby(["School Size"]).mean()["% Passing Reading"]
overall_passing_size = results_df.groupby(["School Size"]).mean()["% Overall Passing"]

# create DataFrame with the above metrics
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": overall_passing_size
})

# display the table
display(size_summary)
```
